# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries

#한수
import matplotlib.pyplot as plt
import numpy as np
import math
# Create an array of x values from -10 to 10
x = np.linspace(0, 10, 100)

# Compute the corresponding y values
y = np.linspace(3, 3,100)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x aixs')
ax.set_ylabel('y aixs')

# Add a title
ax.set_title('$y = x^3$')

# Display the plot
plt.show()
```



(d) $y=-\log_{10}(x)$

    (d-i)(0<x<100)
    
    (d-ii)(0<x<1000 and set x axis to logscale)



```python
import matplotlib.pyplot as plt
import numpy as np
import math
# Create an array of x values from -10 to 10
x = np.logspace(0.1, 10, 100)

# Compute the corresponding y values
y = np.log10(x)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
plt.yscale('log')
ax.set_xlabel('x aixs')
ax.set_ylabel('y aixs')

# Add a title
ax.set_title('$y = x^3$')

# Display the plot
plt.show()
```
