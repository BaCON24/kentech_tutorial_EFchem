# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries

#한수
import matplotlib.pyplot as plt
import numpy as np
import math
# Create an array of x values from -10 to 10
x = np.linspace(0, 10, 100)

# Compute the corresponding y values
y = np.linspace(3, 3,100)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x aixs')
ax.set_ylabel('y aixs')

# Add a title
ax.set_title('$y = x^3$')

# Display the plot
plt.show()
```

    (e-i) Linear x axis, Linear y axis
    
    (e-ii) Log x axis, Linear y axis



```python
import matplotlib.pyplot as plt
import numpy as np
import math
# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)
t= ["t= 1", "t= 10","t= 100"]
# Compute the corresponding y values
y = np.exp(-x)
y1 = np.exp(-x/10)
y2 = np.exp(-x/100)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)
ax.plot(x, y1)
ax.plot(x, y2)
# Add labels for x and y axes
ax.set_xlabel('x aixs')
ax.set_ylabel('y aixs')
# Add a title
ax.set_title('$y = exp(-x/t)$')
# Display the plot
plt.legend(t)
# Create an array of x values from -10 to 10
# Display the plot
plt.show()

x = np.linspace(-10, 10, 100)
t= ["t= 1", "t= 10","t= 100"]
# Compute the corresponding y values
y = np.exp(-x)
y1 = np.exp(-x/10)
y2 = np.exp(-x/100)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)
ax.plot(x, y1)
ax.plot(x, y2)
# Add labels for x and y axes
ax.set_xlabel('x aixs')
ax.set_ylabel('y logscale')
# Add a title
ax.set_title('$y = exp(-x/t)$')
plt.yscale('log')
# Display the plot
plt.legend(t)
# Create an array of x values from -10 to 10
# Display the plot
plt.show()
```
