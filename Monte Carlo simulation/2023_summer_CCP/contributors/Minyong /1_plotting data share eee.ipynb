# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$')

# Display the plot
plt.show()
```

```python
#(a) y=x^2+4*x-5
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,10, 100)
y=x**2+4*x-5
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y=x^2+4*x-5$')

def solve_quadratic(a, b, c):
    #a*x^2+b*x+c
    # Here you define the discriminant, D
    D = b**2- 4*a*c


    # calculate two solutions
    sol1 = (-b - np.sqrt(D)) / (2 * a)
    sol2 = (-b + np.sqrt(D)) / (2 * a)

    return sol1, sol2

a = 1
b = 4
c = -5

sol1, sol2 = solve_quadratic(a, b, c)

ax.plot(sol1,0,"o")
ax.plot(sol2,0,"o")

print(f'The solutions of this quadric are ({sol1},0) and ({sol2},0) ')

plt.show()
```

```python
#(b) y=7*cos(pi*x) (0<x<10)
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=7*np.cos(np.pi*x)
fig, ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y=7cos(pi*x)$')
plt.show()
```

```python
#(d) y=-log10(x)
#d-i
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,100,100)
y=-1*np.log10(x)
fig, ax=plt.subplots(2,1)
ax[0].plot(x,y)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('$y=-log10(x)$')



#d-ii
ax[1].plot(x,y)
ax[1].set_xscale('log')
ax[1].set_xlabel('x-log scale')
ax[1].set_ylabel('y')
ax[1].set_title('$y=-log10(x)$')

plt.tight_layout()
plt.show()
```

```python
#(e)y=exp(-x/t) with t=1,10,100
#e-i Linear x,y axis
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,100,100)
a=[1,10,100]
fig, ax = plt.subplots(2,1)
for t in a :
  y=np.exp(-x/t)
  ax[0].plot(x,y, label=f't={t}')

ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('$y=exp(-x/t)$')


#e-ii Linearx, log y axis
ax[1].set_xlabel('x')
ax[1].set_ylabel('y-logscale')
ax[1].set_title('$y=exp(-x/t)$')

ax[1].set_yscale("log")
for t in a :
  y=np.exp(-x/t)
  ax[1].plot(x,y, label=f't={t}')



ax[0].legend()
ax[1].legend()

plt.tight_layout()
plt.show()


#y=exp(-x^2/D) with D=1,5,10 (-10<x<10)
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-10,10,100)
a=[1,5,10]
fig, ax = plt.subplots(2,1)
for D in a :
  y=np.exp(-x**2/D)
  ax[0].plot(x,y, label=f'D={D}')

ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('$y=exp(-x^2/D)$')

#e-ii Linearx, log y axis
ax[1].set_xlabel('x')
ax[1].set_ylabel('y-logscale')
ax[1].set_title('$y=exp(-x^2/D)$')

ax[1].set_yscale("log")
for D in a :
  y=np.exp(-x**2/D)
  ax[1].plot(x,y, label=f'D={D}')



ax[0].legend()
ax[1].legend()

plt.tight_layout()
plt.show()
```
