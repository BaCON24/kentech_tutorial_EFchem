# 1 - Distribution of random points in 1D

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points(lower_bound, upper_bound, num_points): #최소, 최대, 뽑을 수 개수
    return np.random.uniform(lower_bound, upper_bound, num_points)
    #리턴함, uniform 은 균등 분포, 즉 특정 수를 뽑을 가능성이 모두 같다는 정말 랜덤랜덤하다는 뜻임

# Test the function
# Feed in your numbers
lower_bound = 1  # Lower bound of the int
upper_bound = 6  # Upper bound of the interval
num_points = 100  # Number of random points to generate
random_points = generate_random_points(lower_bound, upper_bound, num_points)

#소수점을 조절할 순 없나?

# print the random points
print(random_points)

# plot the points
plt.figure(figsize=(10, 2))
plt.scatter(random_points, np.zeros_like(random_points), marker='o')
plt.xlim([lower_bound, upper_bound])
plt.yticks([])  # Hide the y-axis
plt.xlabel('x')
plt.title('Random points on a 1D interval')
plt.show()
```

# 2 - Distributing random points in 2D

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points_2D(x_bounds, y_bounds, num_points):
    xv = np.random.uniform(x_bounds[0], x_bounds[1], num_points)
    yv = np.random.uniform(y_bounds[0], y_bounds[1], num_points)
    return xv, yv

# Test the function
# Here is the example of the bounds
x_bounds = [5, 15]  # Bounds for the x-coordinates
y_bounds = [10, 20]  # Bounds for the y-coordinates
num_points = 200  # Number of random points to generate

# Call the function
x_coords, y_coords = generate_random_points_2D(x_bounds, y_bounds, num_points)


# Plot the points
plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, marker='o')
plt.xlim(x_bounds)#x값의 범위는 5-15
plt.ylim(y_bounds)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random points in a 2D interval')
plt.show()
cord = []
for k in range(0, num_points):
  cord.append((x_coords[k], y_coords[k]))
print(cord)

```

# 3 - Find pairs of points with an overlap

Below is an example where some of points overlap (within a certain distance). You will write a function to find those pairs.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points_2D(x_bounds, y_bounds, num_points):
    xv = np.random.uniform(x_bounds[0], x_bounds[1], num_points)
    yv = np.random.uniform(y_bounds[0], y_bounds[1], num_points)
    return xv, yv

def distance(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def plot_connected_points(x_coords, y_coords, max_distance):
    plt.figure(figsize=(6, 6))
    plt.scatter(x_coords, y_coords, marker='o')

    for i in range(len(x_coords)):
        for j in range(i+1, len(x_coords)):
            if distance((x_coords[i], y_coords[i]), (x_coords[j], y_coords[j])) <= max_distance:
                plt.plot([x_coords[i], x_coords[j]], [y_coords[i], y_coords[j]],color = "red", alpha=0.3, linewidth ="2") #알파는 투명도(0~1)

    plt.xlim(x_bounds)
    plt.ylim(y_bounds)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Random points with connected lines')
    plt.show()

# Test the function
x_bounds = [5, 15]
y_bounds = [10, 20]
num_points = 200
max_distance = 0.5  # Maximum distance for connecting points with a line

x_coords, y_coords = generate_random_points_2D(x_bounds, y_bounds, num_points)
plot_connected_points(x_coords, y_coords, max_distance)
```
