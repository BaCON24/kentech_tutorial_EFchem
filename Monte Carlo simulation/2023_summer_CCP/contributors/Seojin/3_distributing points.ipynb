# Part III. Distributing points

# 1 - Distribution of random points in 1D

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points(lower_bound, upper_bound, num_points):
    return np.random.uniform(lower_bound, upper_bound, num_points)

# Test the function
# Feed in your numbers
lower_bound = 5  # Lower bound of the interval
upper_bound = 15  # Upper bound of the interval
num_points = 50  # Number of random points to generate
random_points = generate_random_points(lower_bound, upper_bound, num_points)

# print the random points
print(random_points)

# plot the points
plt.figure(figsize=(10, 2))
plt.scatter(random_points, np.zeros_like(random_points), marker='o')
plt.xlim([lower_bound, upper_bound])
plt.yticks([])  # Hide the y-axis
plt.xlabel('x')
plt.title('Random points on a 1D interval')
plt.show()
```

# 2 - Distributing random points in 2D

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points_2D(x_bounds, y_bounds, num_points):
    # Make your own function
    x_coords=np.random.uniform(x_bounds[0],x_bounds[1],num_points)
    y_coords=np.random.uniform(y_bounds[0],y_bounds[1],num_points)
    return x_coords, y_coords

# Test the function
# Here is the example of the bounds
x_bounds = [5, 15]  # Bounds for the x-coordinates
y_bounds = [10, 20]  # Bounds for the y-coordinates
num_points = 200  # Number of random points to generate

# Call the function
x_coords, y_coords = generate_random_points_2D(x_bounds, y_bounds, num_points)

# Plot the points
plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, marker='o')
plt.xlim(x_bounds)
plt.ylim(y_bounds)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random points in a 2D interval')
plt.show()
```

# 3 - Find pairs of points with an overlap

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points_2D(x_bounds, y_bounds, num_points):
    # Make your own function
    x_coords=np.random.uniform(x_bounds[0],x_bounds[1],num_points)
    y_coords=np.random.uniform(y_bounds[0],y_bounds[1],num_points)
    return x_coords, y_coords
def find_overlapping_pairs(x_coords, y_coords, overlap_distance):
  overlap_pairs=[]
  for i in range(0, num_points):
    x1=x_coords[i]
    y1=y_coords[i]
    for k in range(0, num_points):
      x2=x_coords[k]
      y2=y_coords[k]

      d=np.sqrt((x1-x2)**2+(y1-y2)**2)
      if d<=overlap_distance:
        overlap_pairs.append([(x1,y1),(x2,y2)])
  return overlap_pairs

# Test the function
# Here is the example of the bounds
x_bounds = [5, 15]  # Bounds for the x-coordinates
y_bounds = [10, 20]  # Bounds for the y-coordinates
num_points = 200  # Number of random points to generate

# Call the function
x_coords, y_coords = generate_random_points_2D(x_bounds, y_bounds, num_points)

overlap_distance=0.5
overlap_pairs=find_overlapping_pairs(x_coords, y_coords, overlap_distance)

# Plot the points
plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, marker='o')
for pair in overlap_pairs:
    x = [pair[0][0], pair[1][0]]
    y = [pair[0][1], pair[1][1]]
    plt.plot(x, y, 'red')
plt.xlim(x_bounds)
plt.ylim(y_bounds)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random points in a 2D interval')
plt.show()
```

# 4 - Generating random points without overlaps

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points_2D(x_bounds, y_bounds, num_points, overlap_distance):
    x_coords = []
    y_coords = []

    while len(x_coords) < num_points:
        x1 = np.random.uniform(x_bounds[0], x_bounds[1])
        y1 = np.random.uniform(y_bounds[0], y_bounds[1])

        i = all(np.sqrt((x1 - x)**2 + (y1 - y)**2) > overlap_distance for x, y in zip(x_coords, y_coords))
        if i:
          x_coords.append(x1)
          y_coords.append(y1)

    return x_coords, y_coords

# Test the function
# Here is the example of the bounds
x_bounds = [5, 15]  # Bounds for the x-coordinates
y_bounds = [10, 20]  # Bounds for the y-coordinates
num_points = 200  # Number of random points to generate
overlap_distance=0.5

# Call the function
x_coords, y_coords = generate_random_points_2D(x_bounds, y_bounds, num_points,overlap_distance)

# Plot the points
plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, marker='o')
plt.xlim(x_bounds)
plt.ylim(y_bounds)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random points in a 2D interval')
plt.show()
```
