# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

````python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$')

# Display the plot
plt.show()
````

````python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -5 to 10
x = np.linspace(-5, 10, 100)

# Compute the corresponding y values
y = x ** 2+4*x-5

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)
plt.scatter([-5,1],[0,0])

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y=x^2+4x–5$ (-5<x<10)')

# Display the plot
plt.show()
````

(d) $y=-\log_{10}(x)$


```
#  (d-i)(0<x<100)
    (d-ii)(0<x<1000 and set x axis to logscale)
```



````python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.01, 100, 100)
x2 = np.linspace(0.01, 1000, 100)
y1 = -np.log10(x1)
y2 = -np.log10(x2)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots(2,1)

# Plot the data
ax[0].plot(x1, y1)
ax[1].plot(x2, y2)

# Add labels for x and y axes
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].set_xlabel('x(log scale)')
ax[1].set_ylabel('y')
ax[1].set_xscale('log')

# Add a title
ax[0].set_title('$y=-\log_{10}(x)$ ')
ax[1].set_title('$y=-\log_{10}(x)$ (log scale) ')

# Display the plot
plt.tight_layout()
plt.show()
````

(e) $y = \exp(-x/t)$ with $t$ = 1,10,100

    (e-i) Linear x axis, Linear y axis
    
    (e-ii) Log x axis, Linear y axis

````python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0, 100, 100)
x2 = np.linspace(0, 100, 100)

# Compute the corresponding y values
y1_1 = np.exp(-x1/1)
y1_2 = np.exp(-x1/10)
y1_3 = np.exp(-x1/100)
y2_1 = np.exp(-x2/1)
y2_2 = np.exp(-x2/10)
y2_3 = np.exp(-x2/100)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots(2,1)

# Plot the data
ax[0].plot(x1, y1_1,label='t=1')
ax[0].plot(x1, y1_2,label='t=10')
ax[0].plot(x1, y1_3,label='t=100')
ax[1].plot(x2, y2_1,label='t=1')
ax[1].plot(x2, y2_2,label='t=10')
ax[1].plot(x2, y2_3,label='t=100')

# Add labels for x and y axes
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_xscale('log')

# Add a title
ax[0].set_title('$y = \exp(-x/t)$ with $t$ = 1,10,100 (i)')
ax[1].set_title('$y = \exp(-x/t)$ with $t$ = 1,10,100 (ii)')

# Display the plot
ax[0].legend()
ax[1].legend()
plt.tight_layout()
plt.show()
````

(f) $y = \exp(-x^2/D)$ with $D$ = 1,5,10 ( -10 < x < 10 )

    (Optional) Think about how to present data effectively


````python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y1 = np.exp(-x**2/1)
y2 = np.exp(-x**2/5)
y3 = np.exp(-x**2/10)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y1,label='D=1')
ax.plot(x, y2,label='D=5')
ax.plot(x, y3,label='D=10')

# Add labels for x and y axes
ax.set_xlabel('x(log scale)')
ax.set_ylabel('y')
ax.set_xscale('log')

# Add a title
ax.set_title('$y = \exp(-x^2/D)$ with $D$ = 1,5,10 ( -10 < x < 10 )')

# Display the plot
plt.legend()
plt.show()
````
