# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$')

# Display the plot
plt.show()
```

```python
# import necessary libraries(a)
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -5 to 10
x = np.linspace(-5, 10, 100)

# Compute the corresponding y values
y = x ** 2+x*4-5

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$+x*4-5')

# Display the plot
plt.show()
```

```python
# import necessary libraries(e-i)
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from 0 to 100
x = np.linspace(0, 100, 100)

# Compute the corresponding y values
y1=np.exp(-x/1)
y2=np.exp(-x/10)
y3=np.exp(-x/100)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
line1=ax.plot(x, y1)
line2=ax.plot(x, y2)
line3=ax.plot(x, y3)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = exp(-x/t)with t=1,10,100')

# Display the plot
plt.show()
```
